# Data Inspection, Cleaning, and EDA

In this project, I will build a supervised regression classifier in order to predict medical costs. The variables used are age, sex, bmi, number of children, smoker, and region. The target variable is charges.

### Start with usual imports and data inspection

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../insurance.csv')
df.sample(10)
```

```python
# Examine the sahpe of the data
df.shape
```

```python
# Examine the cardinality of the data
df.describe(include='all')
```

```python
print(len(df['age'].unique()), len(df['bmi'].unique()), len(df['children'].unique()))
```

There are two numerical variables (bmi and age) and one numerical target variable (charges), and four categorical variables (sex, children, smoker, and region).

### Data cleaning

```python
# Check for null values
df.isnull().sum()
```

At first glance it appears that there are no null values.

```python
# Check variable types
df.info()
```

Numerical data appears to already be integer or float types.

There is no need to do any data cleaning.

### Exploratory Data Analysis

Now, I will take a look at each variable in more detail.

```python
# Plot each variable against every other variable to examine relationships. 
# Also look at how each variable is distributed.
sns.pairplot(df)
```

There don't appear to be any correlations. I can see that age is not normally distributed, while bmi is. I will therefore use min-max scaler on both of these numerical variables prior to doing machine learning.

Below, I will take a look at some of the data plotted in more detail.

```python
# Graph the charges
plt.figure(figsize=(10,6))
sns.histplot(df.charges)
```

```python
# Graph charges by smoker.
plt.figure(figsize=(10, 6))
sns.histplot(x=df.charges, hue=df.smoker, element='step')
```

```python
# Graph charges against bmi and smoker.
plt.figure(figsize=(10, 6))
sns.lmplot(x="bmi", y="charges", hue='smoker', data=df, size=8)
```
